# Read UKBiobank data

```python
prs_path = "/oak/stanford/groups/mrivas/projects/multiomics/ukb24983_PRSmap.20211029.phe.gz"
```

```python
def read_data_from_path(path, open_function):
    
    i = 0
    data = []

    with open_function(path, "r") as f:
        for line in f:
            if i % 50000 == 0:
                print('Processed line number ' + str(i))
            i += 1
            line_split = str(line).split('\\t')
            if line_split[1] in iid_dic:
                data.append(line_split)
    
    print()
    print('Finished processing lines.')
    
    return data
```

```python
%%time

data = read_data_from_path(prs_path, gzip.open)
```
